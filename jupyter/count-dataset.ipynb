# データセットの内訳をカウントする

## 1. 必要ライブラリのインポート

```python
from Bio import SeqIO
import os
```

## 2. BIAS-PROFS データセットについてカウント

```python
biasprofs_path = "../data/gds_dataset"
paths = [os.path.join(biasprofs_path, f) for f in os.listdir(biasprofs_path) if f.endswith(".txt")]
```

```python
for path in paths:
    class_name = path.split("Class")[1][0]
    family_name = os.path.basename(path).split(f"Class{class_name}_")[1].replace(".txt", "").replace("_", "/")

    n = 0

    with open(path, "r", encoding="utf-8") as handle:
        for record in SeqIO.parse(handle, "fasta"):
            n += 1

    print(f"| {class_name} | {family_name} | {n} |")
```

```python
class_names = list("ABCDEF")
interpro_path = "../data/interpro"
```

```python
for cls in class_names:
    dir = os.path.join(interpro_path, cls)
    paths = [os.path.join(dir, f) for f in os.listdir(dir) if f.endswith(".fasta")]

    for path in paths:
        family_name = os.path.basename(path).replace(".fasta", "")
        n = 0

        with open(path, "r", encoding="utf-8") as handle:
            for record in SeqIO.parse(handle, "fasta"):
                n += 1

        print(f"| {cls} | {family_name} | {n} |")
```
